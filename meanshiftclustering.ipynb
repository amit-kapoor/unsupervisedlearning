```python
import pandas as pd
import numpy as np
```

```python
titanic_data = pd.read_csv("./data/titanic/train.csv")
titanic_data.head()
```

```python
# drop columns that are specific to passengers
titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], 'columns', inplace=True)
titanic_data.head()
```

```python
from sklearn import preprocessing

laben_enc = preprocessing.LabelEncoder()
titanic_data['Sex'] = laben_enc.fit_transform(titanic_data['Sex'].astype(str))
titanic_data.dtypes
```

```python
# one hot encoding
titanic_data = pd.get_dummies(titanic_data, columns=['Embarked'])
titanic_data.head()
```

```python
# Check for null data
titanic_data[titanic_data.isnull().any(axis=1)]
```

```python
# drom null data
titanic_data = titanic_data.dropna()
```

```python
# estimate bandwidth
from sklearn.cluster import MeanShift, estimate_bandwidth

estimate_bandwidth(titanic_data)
```

```python
# Lets apply it in MeanShift

ms = MeanShift(bandwidth=30)
ms.fit(titanic_data)
```

```python
# find labels
labels = ms.labels_
labels
```

```python
# Lets see unique clusters
np.unique(labels)
```

## 5 unique Clusters

```python
# Include cluster groups now in our titanic data frame

titanic_data['cluster_group'] = np.nan
for i in range(len(titanic_data)):
    titanic_data.iloc[i, titanic_data.columns.get_loc('cluster_group')] = labels[i]

titanic_data.head()
```

```python
# describe data
titanic_data.describe()
```

```python
# Group by Cluster_group
titanic_clustered_data = titanic_data.groupby(['cluster_group']).mean()
```

```python
titanic_clustered_data
```

```python
titanic_clustered_data['Count'] = pd.Series(titanic_data.groupby(['cluster_group']).size())
titanic_clustered_data
```
